```python
# import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage.restoration import denoise_nl_means, estimate_sigma
```

```python
from skimage import img_as_ubyte, img_as_float,io
```

```python
# read image
img=cv2.imread("Sample 5-20 X.tif",0)
fig=plt.figure(figsize=(8,8))
plt.axis("off")
plt.imshow(img)
```

```python
# median_blur-replace values of kernel size-7 by median
median=cv2.medianBlur(img,7)
fig=plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(img,cmap='gray')
```

```python
# blurring an image or averaging the entire pixel values inside the kernel
avging = cv2.blur(img,(15,15))
fig=plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(avging,cmap='gray')
```

```python
# Gaussian_filter-more fine way of blurring
gausBlur = cv2.GaussianBlur(img, (15,15),0)
fig=plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(gausBlur,cmap='gray')
```

```python
# histogram equalizition-to enhance contrast between two colors
from cv2 import equalizeHist
res=equalizeHist(img)
plt.axis('off')
fig=plt.figure(figsize=(8,8))
plt.imshow(res,cmap='gray')
```

```python
# non local means denoising-another pixel-averaging method
float_img=img_as_float(io.imread("Sample 5-10 X.tif"))
```

```python
sigma_est= np.mean(estimate_sigma(float_img, multichannel=True))
denoise_img=denoise_nl_means(float_img, h=1.15 * sigma_est, fast_mode=True,patch_size=5, patch_distance=3, multichannel=True)
denoise_img_as_8byte= img_as_ubyte(denoise_img)
plt.axis("off")
fig=plt.figure(figsize=(8,8))
plt.imshow(denoise_img_as_8byte,cmap='gray')
```
